# PostgreSQL with Python and psycopg2

```python
import psycopg2 as pg2
```

```python
secret = "password"
```

```python
# connect to pgadmin4
conn = pg2.connect(
    database='dvdrental',
    user='postgres',
    password=secret,
    host='localhost',
    port='5432'
)
```

```python
# A pointer to our database
cur = conn.cursor()
```

```python
cur.execute('SELECT * FROM payment')
```

```python
data = cur.fetchmany(2)
```
